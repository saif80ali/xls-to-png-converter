```python
for item in col_names[9:10]:
    dict_options_value = {}
    survey_data = df[item].tolist()
    options_of_each_question = list(set(survey_data))
    
    for option in options_of_each_question:
        dict_options_value[option] = survey_data.count(option)
        
    plt.style.use('ggplot')

    cs=cm.Set1(np.arange(40)/len(dict_options_value))
    plt.figure(figsize=(14,14.5)).add_subplot(111)
    plt.pie(dict_options_value.values(),startangle=90,autopct='%.2f%%',colors=cs)
    
#     bar_plot = plt.bar(dict_options_value.keys(),dict_options_value.values(),color= 'purple')
#     xyz = [f"{'{:.2f}'.format((word/sum(dict_options_value.values()))*100)}%" for word in list(dict_options_value.values())]
#     def autolabel(rects):
#         for idx,rect in enumerate(bar_plot):
#             height = rect.get_height()
#             plt.text(rect.get_x() + rect.get_width()/2., 1*height,
#             xyz[idx],
#             ha='center', va='bottom', rotation=0)
#     autolabel(bar_plot)
    
    plt.title("SolarRide is manufacturing Solar\n Power based Scooties with the\n initiative to boycott fuel driven\n vehicles from India\nAre you happy with it? ", fontdict = font1)
#     plt.xticks(rotation=20)
    
    plt.legend(dict_options_value.keys(),loc=9, bbox_to_anchor=(1,1))
    plt.savefig(item[:10]+'.png')
```
